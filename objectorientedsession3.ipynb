```python
#definind a ciruclar vector field
import matplotlib.pyplot as plt
import numpy as np
```

```python
def circular(x,y):
    u = -y
    v = x
    return (u, v)
```

```python
class VectorField:
    def __init__(self, field):
        self.field = field

    def __call__(self, x:float, y:float): #these define the overall beh of class
        return self.field(x, y)
    
    def plot(self, limits=(-2, 2, -2, 2), scale:int = 10): #these are personalized methods we are using
        X, Y = np.meshgrid(np.linspace(-2, 2, 10), np.linspace(-2, 2, 10)) #2 vectors of 1 dimension, one repeated along x and one repeated along y. all possible combos b/w x and y
        U, V = self.field(X, Y)
        plt.quiver(X, Y, U, V, scale=50)
        plt.show()
```

```python
vf = VectorField(circular)
vf(1, 0)
vf.plot()
```

```python
def bam(x, y):
    u = np.sin(x*y)
    v = np.tan(x+y)
    return u,v
```

```python
vf2 = VectorField(bam)
vf2.plot()
```
