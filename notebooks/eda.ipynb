### data cleaning and processing

```python
medalsdata["medal"].value_counts() #if empty column, that means they did not get a medal
```

filling empty columns:

```python
medalsdata["medal"].fillna("No", inplace = True)
```

```python
medalsdata["medal"].value_counts()
```

splitting columns

```python
#we want to split the edition column into year and edition
medalsdata["edition"]
```

```python
col = medalsdata["edition"].str.split(" ").str[0] #split finds the first space and splits the string there. then treat this as a string and take the first element, which is the year
col
```

```python
medalsdata["year"] = col
medalsdata
```

```python
#now we look at the season:
col = medalsdata["edition"].str.split(" ").str[1]
col
```

```python
medalsdata["season"] = col
medalsdata
```

```python
#finding the gender
medalsdata["gender"] = medalsdata["event"].str.split(",").str[-1]
medalsdata
```

```python
medalsdata["gender"].value_counts()
```

```python
#genderdict = {"Women":"Female", "Men":"Male", "Boys":"Male", "Open":"Open", "Mixed":"Mixed"}
#medalsdata["gender"] = medalsdata["gender"].map(genderdict)
```

```python
medalsdata.drop(columns = "pos", inplace = True)
medalsdata
```

```python
#athlete with the most medals:
allmedals = medalsdata[medalsdata["medal"] != "No"]
allmedals["athlete"].value_counts()
```

```python
#count number of gold medals won by each country in 2020.
medals2020 = medalsdata[medalsdata["year"] == "2020"]
onlymedals2020 = medals2020[medals2020["medal"] != "No"]
onlygoldmedals2020 = onlymedals2020[onlymedals2020["medal"] == "Gold"]
onlygoldmedals2020["country_noc"].value_counts()
```

```python
sorted = onlygoldmedals2020[["country_noc", "medal"]].groupby(by = "country_noc", as_index = False).count().sort_values(by = "medal", ascending = False)
top10goldmedals = sorted.iloc[0:10, :]
top10goldmedals.reset_index(inplace = True)
top10goldmedals.drop(columns = "index", inplace = True)
top10goldmedals
```
