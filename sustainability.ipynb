```python
data2010 = pd.read_csv("historic_demand_year_2010.csv")
data2011 = pd.read_csv("historic_demand_year_2011.csv")
data2012 = pd.read_csv("historic_demand_year_2012.csv")
data2013 = pd.read_csv("historic_demand_year_2013.csv")
data2014 = pd.read_csv("historic_demand_year_2014.csv")
data2015 = pd.read_csv("historic_demand_year_2015.csv")
data2016 = pd.read_csv("historic_demand_year_2016.csv")
```

```python
ls2009 = data2009.columns.to_list()
ls2009
```

```python
for i in ls2009:
    data2009[i.upper()] = data2009[i]
    data2009.drop(columns = i)

data2009
```

```python
data2009_2016 = pd.concat([data2009, data2011, data2012, data2013, data2014, data2015, data2016])
data2009_2016
```

```python
data2009_2016["SETTLEMENT_DATE"] = pd.to_datetime(data2009_2016["SETTLEMENT_DATE"])
```

```python
data2009_2016["YEAR"] = data2009_2016["SETTLEMENT_DATE"].dt.year
data2009_2016["MONTH"] = data2009_2016["SETTLEMENT_DATE"].dt.month
data2009_2016
```

```python
avgtsd_yearmonthperiod = data2009_2016.loc[:, ["SETTLEMENT_PERIOD", "TSD", "YEAR", "MONTH"]].groupby(by = ["SETTLEMENT_PERIOD", "YEAR", "MONTH"], as_index=False).mean()
avgtsd_yearmonthperiod
```

```python
avgtsd_yearmonthperiod["PERIOD_YEAR_MONTH"] = avgtsd_yearmonthperiod["SETTLEMENT_PERIOD"].astype(str) + "-" + avgtsd_yearmonthperiod["YEAR"].astype(str) + "-" + avgtsd_yearmonthperiod["MONTH"].astype(str)
```

```python
avgtsd_yearmonthperiod
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(avgtsd_yearmonthperiod["PERIOD_YEAR_MONTH"], avgtsd_yearmonthperiod["TSD"])
```

```python
avgtsd_yearperiod = avgtsd_yearmonthperiod.loc[:, ["SETTLEMENT_PERIOD", "YEAR", "TSD"]].groupby(by = ["SETTLEMENT_PERIOD", "YEAR"], as_index = False).mean("TSD")
avgtsd_yearperiod
```

```python
avgtsd_yearperiod["PERIOD-YEAR"] = avgtsd_yearperiod["SETTLEMENT_PERIOD"].astype(str) + "-" + avgtsd_yearperiod["YEAR"].astype(str)
avgtsd_yearperiod
```

```python
plt.plot(avgtsd_yearperiod["PERIOD-YEAR"], avgtsd_yearperiod["TSD"])
```

```python
ls2024 = data2024.columns.to_list()
for i in ls2024:
    data2024[i.upper()] = data2024[i]
    data2024.drop(columns = i, inplace = True)
data2024
```

```python
data2017_2024 = pd.concat([data2017, data2018, data2019, data2020, data2021, data2022, data2023, data2024])
data2017_2024
```

```python
data2017_2024["SETTLEMENT_DATE"] = pd.to_datetime(data2017_2024["SETTLEMENT_DATE"])
```

```python
alldata = pd.concat([data2009_2016, data2017_2024])
alldata
```

```python
meantsdperyear = alldata.loc[:, ["TSD", "YEAR"]].groupby(by = "YEAR", as_index = False).mean()
meantsdperyear
```

```python
plt.plot(meantsdperyear["YEAR"], meantsdperyear["TSD"])
```

```python
#creating a line to fit this data
X = np.array(meantsdperyear["YEAR"])
Y = np.array(meantsdperyear["TSD"])
```

```python
(np.linalg.inv(X @ X.T)) @ X.T @ Y
```
